```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Uploadind data frame with number of cars in Poland
```

```python
cars = pd.read_excel('cars.xlsx')
cars
```

```python
cars = cars.transpose()
cars
```

```python
cars.reset_index(inplace= True)
cars
```

```python
cars = px.bar(cars,x='index',y=0, title = 'Number of cars in Poland in 1999-2020', labels={'index':'Years', '0':'Number of cars'})
cars.write_html('cars.html')
cars.show()
```

```python
car_accf = pd.read_excel('wypadki.xlsx')
car_accf
```

```python
# NUMBER OF PEOPLE IN POLAND
```

```python
population_pl = px.bar(car_accf,x='Rok',y='Ludnosc',title = 'Number of people in Poland in 1999-2020', labels={'Rok':'Year', 'Ludnosc':'Number of people'})
population_pl.write_html('population_pl.html')
population_pl.show()
```

```python
# Number of car accidents
```

```python
car_acc = px.bar(car_accf,x='Rok',y='wypadki',title = 'Number of car accidents in Poland in 1999-2020', labels={'Rok':'Year', 'wypadki':'Number of car accidents'})
car_acc.write_html('car_acc.html')
car_acc.show()
```

```python
# Number of death car accidents
```

```python
car_acc_df = pd.read_excel('smiertelne.xlsx')
car_acc_df
```

```python
car_acc_d = px.bar(car_acc_df,x='Rok',y='wypadki', title = 'Number of death car accidents in Poland in 1999-2020', labels={'Rok':'Year', 'wypadki':'Number of death car accidents'})
car_acc_d.write_html('car_acc_d.html')
car_acc_d.show()
```

```python
car_acc_join = car_accf.merge(car_acc_df, on='Rok')
car_acc_join
```

```python
car_acc_join['death_percent [%]'] =car_acc_join['wypadki_y']/car_acc_join['wypadki_x']*100
car_acc_join
```

```python
death_percent = px.line(car_acc_join,x='Rok',y='death_percent [%]',title = 'Number of percent of death car accidents in Poland in 1999-2020', labels={'Rok':'Year', 'death_percent [%]':'Death car accidents / Car accidents [%]'})
death_percent.write_html('death_percent.html')
death_percent.show()
```
